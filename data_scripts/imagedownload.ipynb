```python
import os
import pandas as pd
import requests
import json


# Define the path to the CSV file
csv_path = '/content/drive/MyDrive/Cleaned_Indian_Food_Dataset.csv'

# Read the CSV file using pandas
data = pd.read_csv(csv_path)

# Create a directory in Google Drive to save images
output_dir = '/content/drive/MyDrive/downloaded_images'
os.makedirs(output_dir, exist_ok=True)


image_mapping = {}

for index, row in data.iterrows():
    image_url = row['image-url']
    sequential_id = str(index).zfill(4)
    image_name = f"{sequential_id}.jpg"
    image_path = os.path.join(output_dir, image_name)

    # Download the image
    response = requests.get(image_url, stream=True)
    if response.status_code == 200:
        with open(image_path, 'wb') as file:
            file.write(response.content)
        print(f"Downloaded {image_name}")

        image_mapping[sequential_id] = {'imageUrl': image_url}
    else:
        print(f"Failed to download {image_name}")

mapping_json_path = '/content/drive/MyDrive/image_mapping.json'
with open(mapping_json_path, 'w') as json_file:
    json.dump(image_mapping, json_file, indent=4)

print("Image mapping saved as JSON.")
```
