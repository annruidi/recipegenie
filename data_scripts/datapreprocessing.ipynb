```python
import random

counter = 0

def generate_unique_hex_id(length=8):
    global counter
    max_val = 16**(length-4) - 1
    random_num = random.randint(0, max_val)
    counter += 1
    unique_part = counter % (16**4)  # The last 4 chars will be unique
    hex_id = hex((random_num << 16) + unique_part)[2:]
    return hex_id.zfill(length)

print(generate_unique_hex_id())
```

```python
import csv
import json

csv_file_path = '/content/drive/MyDrive/Cleaned_Indian_Food_Dataset.csv'
json_path = '/content/drive/MyDrive/layer1.json'
json_path2 = '/content/drive/MyDrive/layer2.json'
det_ingrs = '/content/drive/MyDrive/det_ingrs.json'

data = []
data2 = []
ingrdata = []

with open(csv_file_path, 'r', encoding='utf-8') as csvfile:
    csvreader = csv.DictReader(csvfile)

    for row in csvreader:


        item = {
            "ingredients": [{"text": ingredient.strip()} for ingredient in row['TranslatedIngredients'].split(',')],
            "url": row['URL'],
            "partition": "train",
            "title": row['TranslatedRecipeName'],
            "id": generate_unique_hex_id(),
            "instructions": [{"text": instruction.strip()} for instruction in row['TranslatedInstructions'].split('.')]
        }



        item2 = {
            "id": item['id'],
            "images": [{
                "id":"",
                "url":row["image-url"]
            }]
        }

        true_ingredients = row['Cleaned-Ingredients'].split(',')
        ingredients = [{"text": ingredient.strip()} for ingredient in true_ingredients]
        valid = [True] * len(true_ingredients)

        item3 = {
            "valid": valid,
            "id": item['id'],
            "ingredients": ingredients
        }


        data.append(item)
        data2.append(item2)
        ingrdata.append(item3)




with open(json_path, 'w', encoding='utf-8') as jsonfile:
    json.dump(data, jsonfile, ensure_ascii=False, indent=4)

with open(json_path2, 'w', encoding='utf-8') as jsonfile:
    json.dump(data2, jsonfile, ensure_ascii=False, indent=4)

with open(det_ingrs, 'w', encoding='utf-8') as jsonfile:
    json.dump(ingrdata, jsonfile, ensure_ascii=False, indent=4)
```
